```python
import pandas as pd
from sklearnex import patch_sklearn
patch_sklearn()
from pycaret.classification import *
```

```python
train = pd.read_csv(r'balanced_train.csv')
train
```

```python
train['Stroke'].value_counts()
```

```python
test = pd.read_csv(r'test.csv')
test
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif
```

```python
xtrain = train.drop('Stroke', axis=1)
ytrain = train['Stroke']
```

```python
selector = SelectKBest(mutual_info_classif, k = 10)
X_new = selector.fit_transform(xtrain, ytrain)
```

```python
selected_features = xtrain.columns[selector.get_support()]
print(selected_features)
```

```python
a = list(selected_features)
a.append('Stroke')
```

```python
train = train.loc[:,a] 
train
```

```python
test = test.loc[:,a]
test
```

```python
test['Stroke'].value_counts()
```

```python
exp_name = setup(
    data = train,  
    target = 'Stroke',
    preprocess = False,
    fold_shuffle = True,
    use_gpu = True,
    session_id = 420
    )
```
