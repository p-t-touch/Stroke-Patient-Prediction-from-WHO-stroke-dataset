```python
x = pd.get_dummies(df['gender'])
df = df.drop('gender', axis = 1)
df = df.join(x)
df
```

```python
x = pd.get_dummies(df['hypertension'])
df = df.drop('hypertension', axis = 1)
df = df.join(x)
df.rename(columns={df.columns[11]: "no_hypertension", df.columns[12]: "yes_hypertension"}, inplace = True)
df
```

```python
x = pd.get_dummies(df['heart_disease'])
df = df.drop('heart_disease', axis = 1)
df = df.join(x)
df.rename(columns={df.columns[12]: "no_heart_disease", df.columns[13]: "yes_heart_disease"}, inplace = True)
df
```

```python
x = pd.get_dummies(df['ever_married'])
df = df.drop('ever_married', axis = 1)
df = df.join(x)
df.rename(columns={df.columns[13]: "never_married", df.columns[14]: "ever_married"}, inplace = True)
df
```

```python
x = pd.get_dummies(df['work_type'])
df = df.drop('work_type', axis = 1)
df = df.join(x)
#df.rename(columns={df.columns[14]: "never_married", df.columns[15]: "ever_married"}, inplace = True)
df
```

```python
x = pd.get_dummies(df['Residence_type'])
df = df.drop('Residence_type', axis = 1)
df = df.join(x)
#df.rename(columns={df.columns[14]: "never_married", df.columns[15]: "ever_married"}, inplace = True)
df
```

```python
x = pd.get_dummies(df['smoking_status'])
df = df.drop('smoking_status', axis = 1)
df = df.join(x)
df.rename(columns={'Unknown': "Unknown_smoking"}, inplace = True)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df.drop(columns = 'stroke'), df[['stroke']], test_size = 0.3, shuffle = True, stratify = df[['stroke']], random_state = 420)
```

```python
xtrain.info()
xtrain
```

```python
xtrain['Stroke'] = ytrain
xtrain.info()
xtrain.reset_index(drop = True, inplace = True)
xtrain
```

```python
xtest['Stroke'] = ytest
xtest.info()
xtest.reset_index(drop = True, inplace = True)
xtest
```

```python
scaler = MinMaxScaler()
xtrain = pd.DataFrame(scaler.fit_transform(xtrain), columns = xtrain.columns)
xtest = pd.DataFrame(scaler.transform(xtest), columns = xtest.columns)
```

```python
xtrain.to_csv('pre_balanced_train.csv', index = False)
xtest.to_csv('test.csv', index = False)
```
